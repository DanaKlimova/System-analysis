### <center> [Вариант 9](http://archive.ics.uci.edu/ml/datasets/Abalone) </center>

```python
import bs4
import numpy as np
import matplotlib.pyplot as plt

from scipy import stats
from urllib.request import urlopen 

url = "http://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data"
retrieved_data = urlopen(url).read()
soup = bs4.BeautifulSoup(retrieved_data)
soup.prettify()
str_soup = str(soup)
data_list = str_soup.split('\n')
data_list.remove("")
x_list = []
y_list = []
for x in data_list:
    a = x.split(',')
    x_list.append(float(a[1]))
    y_list.append(float(a[2]))

slope, intercept, r_value, p_value, std_err = stats.linregress(x_list, y_list)
print("slope: %f" % slope)
print("intercept: %f" % intercept)
print("correlation coefficient: %f" % r_value)
print("y = (%f) * x + (%f)" % (slope, intercept))

plt.plot(x_list, y_list, 'o', label='Abalone Data Set ')
plt.plot(x_list, intercept + slope*np.asarray(x_list), 'm', label='fitted line')
plt.legend()
plt.show()
```
